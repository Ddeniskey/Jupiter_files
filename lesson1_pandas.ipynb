```python
import pandas as pd
bookings = pd.read_csv('https://stepik.org/media/attachments/lesson/360344/bookings.csv', sep = ';')
```

```python
bookings_head = bookings.head(7)
```

```python
bookings.rename(str.lower, axis='columns')
```

```python
bookings.columns = bookings.columns.str.replace(' ', '_',regex = True)
```

```python
bookings = bookings.rename(str.lower, axis='columns')
```

```python
bookings
```

```python
bookings.query('is_canceled == 0').groupby('country', as_index=False). \
                agg({'is_canceled':'value_counts'}). \
                sort_values('is_canceled', ascending = False)
```

```python
bookings.groupby('hotel', as_index = False). \
                agg({'stays_total_nights':'mean'}).round(2)
```

```python
bookings[['reserved_room_type', 'assigned_room_type']]
```

```python
bookings.query('reserved_room_type != assigned_room_type').shape
```

```python
bookings.groupby('arrival_date_year').arrival_date_month.agg(pd.Series.mode)
```

```python
bookings.query("is_canceled == 1 & hotel == 'City Hotel'").groupby('arrival_date_year').arrival_date_month. \
                agg(pd.Series.mode)
```

```python
bookings[['children', 'babies', 'adults']].mean()
```

```python
kids_count = bookings['babies'] + bookings['children']
bookings['total_kids'] = kids_count
```

```python
bookings.groupby('hotel', as_index = False).total_kids.agg(pd.Series.mean).round(2)
```

```python
bookings['has_kids'] = bookings.total_kids != 0
```

```python
df = bookings[['has_kids', 'is_canceled']].value_counts()
```
