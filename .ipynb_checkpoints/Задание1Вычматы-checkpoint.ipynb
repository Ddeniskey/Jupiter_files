```python
#Решение уравнения гиперболического типа схемой Лакса(4ая схема)
a = 0.1 #коэф в уравнении
L = 1 #значение правой границы х
T = 1 # граница времени
n = 5000 # число шагов по времени
m = 100 # число шагов по пространству
h = L / m # шаг по х
t = T / n # шаг по времени

# зададим сетку 
x_i = np.arange(0, L, h) # значения в узлах по х
t_j = np.arange(0, T, t) # значение в узлах по t
r_j = len(t_j) # количество узлов по t
r_i = len(x_i) # количество узлов по x
g_t_h = np.zeros([r_j, r_i]) # итоговая сетка размером t_j*x_i

#начальные условия
def fi(x, u1, u0):
    ans = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] == 0:
            ans[i] = u1
        else: 
            ans[i] = u0
    return(ans)
u1 , u0= 2, 1
u_t_0 = fi(x_i, u1, u0) # значение при t = 0
u_x_0 = u1 # значение при x = 0
g_t_h[0] = u_t_0 # инициализация сетки t = 0
for i in range(n):
    g_t_h[i] = u_t_0 # граничное условие при х = 0
print(len(g_t_h[2999]))
# посчитаем значения сетки со временем
sigm = a * t / h #число Курента
for i in range(1, r_i -1): #пространство
    for j in range(r_j - 1):#время
        g_t_h[j + 1][i] = 0.5 * (g_t_h[j][i + 1] + g_t_h[j][i - 1]) -0.5 * sigm * (g_t_h[j][i+1] - g_t_h[j][i-1])
        #взято из левого уголка,чтобы крайнюю точку по х посчитать
        g_t_h[j +1][r_i-1] = g_t_h[j][r_i-1] - sigm * (g_t_h[j][r_i-1] - g_t_h[j][r_i -2])
print(g_t_h[2999])
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set(xlabel='x', ylabel='u', title='Схема Лакса')
for j in range(r_j):
    ax.plot(x_i, g_t_h[j]) #двумерный график,для каждого х рисуются значения при времени от 0 до T с шагом t
fig3d = plt.figure()
ax3d = fig3d.add_subplot(111, projection = "3d")
ax3d.set(xlabel = "x", ylabel = 'u', zlabel = 't', title = "Лакс в 3d")
for j in range(r_j):
    ax3d.plot(x_i, g_t_h[j], j / n)
```

```python
#Схема Лакса-Вендроффа с монотонизацией
a = 1 #коэф в уравнении
L = 1 #значение правой границы х
T = 1 # граница времени
n = 3000 # число шагов по времени
m = 2000 # число шагов по пространству
h = L / m # шаг по х
t = T / n # шаг по времени
q = 0.25 #параметр монотонизации

# зададим сетку 
x_i = np.arange(0, L, h) # значения в узлах по х
t_j = np.arange(0, T, t) # значение в узлах по t
r_j = len(t_j) # количество узлов по t
r_i = len(x_i) # количество узлов по x
g_t_h = np.zeros([r_j, r_i]) # итоговая сетка размером t_j*x_i

#начальные условия
def fi(x, u1, u0):
    ans = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] == 0:
            ans[i] = u1
        else: 
            ans[i] = u0
    return(ans)
u1 , u0= 2, 1
u_t_0 = fi(x_i, u1, u0) # значение при t = 0
u_x_0 = u1 # значение при x = 0
g_t_h[0] = u_t_0 # инициализация сетки t = 0
for i in range(n):
    g_t_h[i] = u_t_0 # граничное условие при х = 0
print(len(g_t_h[2999]))
# посчитаем значения сетки со временем
sigm = a * t / h #число Курента
for i in range(2, r_i -2): #пространство
    for j in range(r_j - 1):#время
        y_plus = 0.5 * sigm * (sigm - 1) * (g_t_h[j][i+1] - g_t_h[j][i])
        y_minus = 0.5 * sigm * (sigm - 1) * (g_t_h[j][i] - g_t_h[j][i-1])
        g_t_h[j + 1][i] = g_t_h[j][i] - sigm*(g_t_h[j][i] - g_t_h[j][i - 1]) - (y_plus - y_minus)
        #монотонизация
        Dp = g_t_h[j][i + 1] - g_t_h[j][i]
        Dpp = g_t_h[j][i + 2] - g_t_h[j][i +1]
        Dm = g_t_h[j][i] - g_t_h[j][i - 1]
        Dmm = g_t_h[j][i - 1] - g_t_h[j][i - 2]
        if (Dpp * Dp > 0) or (Dp * Dm < 0):
            Q_plus = Dp
        else:
            Q_plus = 0
        if (Dmm * Dm > 0) or (Dp * Dm < 0):
            Q_minus = Dm
        else:
            Q_minus = 0
        #smooth
        g_t_h[j + 1][i] += q * (Q_plus - Q_minus)
        
print(g_t_h[2999])
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set(xlabel='x', ylabel='u', title='Схема Лакса-Вендроффа')
for j in range(r_j):
    ax.plot(x_i, g_t_h[j]) #двумерный график,для каждого х рисуются значения при времени от 0 до T с шагом t
fig3d = plt.figure()
ax3d = fig3d.add_subplot(111, projection = "3d")
ax3d.set(xlabel = "x", ylabel = 'u', zlabel = 't', title = "Лакс-Вендрофф в 3d")
for j in range(r_j):
    ax3d.plot(x_i, g_t_h[j], j / n)
```

```python
#Схема 7 TVD при a > 0
a = 1 #коэф в уравнении
L = 1 #значение правой границы х
T = 1 # граница времени
n = 3000 # число шагов по времени
m = 2000 # число шагов по пространству
h = L / m # шаг по х
t = T / n # шаг по времени
q = 0.05 #параметр монотонизации
eps = 10 ** (-6)

# зададим сетку 
x_i = np.arange(0, L, h) # значения в узлах по х
t_j = np.arange(0, T, t) # значение в узлах по t
r_j = len(t_j) # количество узлов по t
r_i = len(x_i) # количество узлов по x
g_t_h = np.zeros([r_j, r_i]) # итоговая сетка размером t_j*x_i

#начальные условия
def fi(x, u1, u0):
    ans = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] == 0:
            ans[i] = u1
        else: 
            ans[i] = u0
    return(ans)
u1 , u0= 2, 1
u_t_0 = fi(x_i, u1, u0) # значение при t = 0
u_x_0 = u1 # значение при x = 0
g_t_h[0] = u_t_0 # инициализация сетки t = 0
for i in range(n):
    g_t_h[i] = u_t_0 # граничное условие при х = 0
print(len(g_t_h[2999]))
# посчитаем значения сетки со временем
sigm = a * t / h #число Курента
for i in range(2, r_i -2): #пространство
    for j in range(r_j - 1):#время
        r_m = (g_t_h[j][i] - g_t_h[j][i-1] + eps) / (g_t_h[j][i + 1] - g_t_h[j][i] + eps)
        r_m_1 = (g_t_h[j][i-1] - g_t_h[j][i-2] + eps) / (g_t_h[j][i] - g_t_h[j][i-1] + eps)
        if r_m > 1:
            fi_rm = min(2, r_m)
        elif r_m > 0 and r_m <=1:
            fi_rm = min(1, 2*r_m)
        else:
            fi_rm = 0
        if r_m_1 > 1:
            fi_rm1 = min(2, r_m_1)
        elif r_m_1 > 0 and r_m_1 <=1:
            fi_rm1 = min(1, 2*r_m_1)
        else:
            fi_rm1 = 0
        
        y_plus = 0.5 * fi_rm * sigm * (sigm - 1) * (g_t_h[j][i+1] - g_t_h[j][i])
        y_minus = 0.5 *fi_rm1 * sigm * (sigm - 1) * (g_t_h[j][i] - g_t_h[j][i-1])
        g_t_h[j + 1][i] = g_t_h[j][i] - sigm*(g_t_h[j][i] - g_t_h[j][i - 1]) - (y_plus - y_minus)
        #монотонизация
        Dp = g_t_h[j][i + 1] - g_t_h[j][i]
        Dpp = g_t_h[j][i + 2] - g_t_h[j][i +1]
        Dm = g_t_h[j][i] - g_t_h[j][i - 1]
        Dmm = g_t_h[j][i - 1] - g_t_h[j][i - 2]
        if (Dpp * Dp > 0) or (Dp * Dm < 0):
            Q_plus = Dp
        else:
            Q_plus = 0
        if (Dmm * Dm > 0) or (Dp * Dm < 0):
            Q_minus = Dm
        else:
            Q_minus = 0
        #smooth
        g_t_h[j + 1][i] += q * (Q_plus - Q_minus)
        
print(g_t_h[2999])
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set(xlabel='x', ylabel='u', title='Схема Лакса-Вендроффа')
for j in range(r_j):
    ax.plot(x_i, g_t_h[j]) #двумерный график,для каждого х рисуются значения при времени от 0 до T с шагом t
fig3d = plt.figure()
ax3d = fig3d.add_subplot(111, projection = "3d")
ax3d.set(xlabel = "x", ylabel = 'u', zlabel = 't', title = "Лакс-Вендрофф в 3d")
for j in range(r_j):
    ax3d.plot(x_i, g_t_h[j], j / n)
```
