```python
import numpy as np
import scipy
from scipy import special
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
plt.style.use('seaborn-pastel')
```

```python
#Ð½ÐµÐ»Ð¸Ð½ÐµÐ¹Ð½Ð°Ñ ÑÑ…ÐµÐ¼Ð° 2
#ÑÑ…ÐµÐ¼Ð° 2 Ñ‚Ñ€ÐµÑ…ÑÑ‚ÑƒÐ¿ÐµÐ½Ñ‡Ð°Ñ‚Ð°Ñ Ð´Ð»Ñ Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾Ð³Ð¾ ÑƒÑ€Ð°Ð²Ð½ÐµÐ½Ð¸Ñ Ð¿Ñ€Ð¾Ð³Ð¾Ð½ÐºÐ¾Ð¹ Ð³Ñ€Ð°Ð½ ÑƒÑÐ»Ð¾Ð²Ð¸Ñ Ð¿ÐµÑ€Ð²Ð¾Ð³Ð¾ Ñ€Ð¾Ð´Ð°
lambd = 0.1
Temp=1
sigm = 0.1
L = 1
T = 1
C_p = 0.4
m =100
h = L / m
t = C_p * h ** 2 / lambd
coef = t /(h ** 2)
n = int(T // t)
eps = 0.01
x_i = np.arange(0, L+h, h) # Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ Ð² ÑƒÐ·Ð»Ð°Ñ… Ð¿Ð¾ Ñ…
t_j = np.linspace(0, T, n) # Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ð² ÑƒÐ·Ð»Ð°Ñ… Ð¿Ð¾ t
r_j = len(t_j) # ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ ÑƒÐ·Ð»Ð¾Ð² Ð¿Ð¾ t
r_i = len(x_i) # ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ ÑƒÐ·Ð»Ð¾Ð² Ð¿Ð¾ x
g_t_h = np.zeros([r_j, r_i]) # Ð¸Ñ‚Ð¾Ð³Ð¾Ð²Ð°Ñ ÑÐµÑ‚ÐºÐ° Ñ€Ð°Ð·Ð¼ÐµÑ€Ð¾Ð¼ t_j*x_i, Ð¿Ñ€Ð¾ÑÑ‚Ð¾ Ñ‚ÐµÐ¼Ð¿ÐµÑ€Ð°Ñ‚ÑƒÑ€Ð°, t
g1_t_h = np.zeros([r_j, r_i]) # Ñ‚ Ñ Ð²Ð¾Ð»Ð½Ð¾Ð¹ t
g2_t_h = np.zeros([r_j, r_i]) # Ñ‚ Ñ Ñ‡ÐµÑ€Ñ‚Ð¾Ð¹ t / 2
g3_t_h = np.zeros([r_j, r_i]) # Ñ‚ Ñ ÐºÑ€Ñ‹ÑˆÐºÐ¾Ð¹ t / 2
def fi(x, Temp):
    ans = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] == 0:
            ans[i] = Temp
        else: 
            ans[i] = eps
    return(ans)
T_t_0 = fi(x_i, Temp) # Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ð¿Ñ€Ð¸ t = 0
T_x_0 = Temp # Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ð¿Ñ€Ð¸ x = 0
T_x_L = 0
g_t_h[0] = T_t_0 # Ð¸Ð½Ð¸Ñ†Ð¸Ð°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ ÑÐµÑ‚ÐºÐ¸ t = 0
g1_t_h[0] = T_t_0 # Ð¸Ð½Ð¸Ñ†Ð¸Ð°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ ÑÐµÑ‚ÐºÐ¸ t = 0
g2_t_h[0] = T_t_0 # Ð¸Ð½Ð¸Ñ†Ð¸Ð°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ ÑÐµÑ‚ÐºÐ¸ t = 0
g3_t_h[0] = T_t_0 # Ð¸Ð½Ð¸Ñ†Ð¸Ð°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ ÑÐµÑ‚ÐºÐ¸ t = 0
for i in range(n):
    g_t_h[i][0] = T_x_0 # Ð³Ñ€Ð°Ð½Ð¸Ñ‡Ð½Ð¾Ðµ ÑƒÑÐ»Ð¾Ð²Ð¸Ðµ Ð¿Ñ€Ð¸ Ñ… =0
    g1_t_h[i][0] = T_x_0 # Ð³Ñ€Ð°Ð½Ð¸Ñ‡Ð½Ð¾Ðµ ÑƒÑÐ»Ð¾Ð²Ð¸Ðµ Ð¿Ñ€Ð¸ Ñ… =0
    g2_t_h[i][0] = T_x_0 # Ð³Ñ€Ð°Ð½Ð¸Ñ‡Ð½Ð¾Ðµ ÑƒÑÐ»Ð¾Ð²Ð¸Ðµ Ð¿Ñ€Ð¸ Ñ… =0
    g_t_h[i][r_i-1] = 0 # Ð³Ñ€Ð°Ð½Ð¸Ñ‡Ð½Ð¾Ðµ ÑƒÑÐ»Ð¾Ð²Ð¸Ðµ Ð¿Ñ€Ð¸ Ñ… =0
    g1_t_h[i][r_i-1] = 0 # Ð³Ñ€Ð°Ð½Ð¸Ñ‡Ð½Ð¾Ðµ ÑƒÑÐ»Ð¾Ð²Ð¸Ðµ Ð¿Ñ€Ð¸ Ñ… =0
    g2_t_h[i][r_i-1] = 0 # Ð³Ñ€Ð°Ð½Ð¸Ñ‡Ð½Ð¾Ðµ ÑƒÑÐ»Ð¾Ð²Ð¸Ðµ Ð¿Ñ€Ð¸ Ñ… =0
    g3_t_h[i][r_i-1] = 0 # Ð³Ñ€Ð°Ð½Ð¸Ñ‡Ð½Ð¾Ðµ ÑƒÑÐ»Ð¾Ð²Ð¸Ðµ Ð¿Ñ€Ð¸ Ñ… =0
#print(g_t_h)
alph1 = np.zeros([r_i-1])
beta1 = np.zeros([r_i-1])
alph2 = np.zeros([r_i-1])
beta2 = np.zeros([r_i-1])
alph3 = np.zeros([r_i-1])
beta3 = np.zeros([r_i-1])
lambd_plus = np.zeros([r_i-1])
lambd_minus = np.zeros([r_i-1])
for j in range(r_j-1):
    alph1[0] = 0
    beta1[0] = g_t_h[j+1][0]
    alph2[0] = 0
    beta2[0] = g_t_h[j+1][0]
    alph3[0] = 0
    for i in range(1, r_i-1): #Ð²Ñ‹Ñ‡Ð¸ÑÐ»ÑÐµÐ¼ ÐºÐ¾ÑÑ„Ñ„Ð¸Ñ†Ð¸ÐµÐ½Ñ‚Ñ‹ Ð°Ð»ÑŒÑ„Ð° Ð¸ Ð±ÐµÑ‚Ð° Ð´Ð»Ñ g1
        lambd_i_1 = lambd * (g_t_h[j][i - 1])**sigm
        lambd_i = lambd * (g_t_h[j][i])**sigm
        lambd_i1 = lambd * (g_t_h[j][i+1])**sigm
        lambd_plus[i] = 2 * (lambd_i * lambd_i1) / (lambd_i + lambd_i1)
        lambd_minus[i] = 2 * (lambd_i*lambd_i_1) / (lambd_i + lambd_i_1)
        alph1[i] = C_p / (-C_p* alph1[i - 1] + 1 + 2 * C_p)
        beta1[i] = (g_t_h[j][i] + C_p * beta1[i-1]) / (-C_p * alph1[i-1] + 1 + 2 * C_p)
        alph2[i] = 0.5*C_p / (-0.5*C_p* alph2[i - 1] + 1 + C_p)
        beta2[i] = (g_t_h[j][i] + 0.5*C_p * beta2[i-1]) / (-0.5*C_p * alph2[i-1] + 1 + C_p)   
    #Ð³Ñ€Ð°Ð½ ÑƒÑÐ»Ð¾Ð²Ð¸Ðµ Ð´Ð»Ñ Ð¿ÐµÑ€Ð²Ð¾Ð³Ð¾ Ñ€Ð¾Ð´Ð°
    g1_t_h[j+1][r_i-1] = (g_t_h[j][r_i-1] + coef * lambd_minus[r_i-2] * beta1[r_i-2]) / (1 + (lambd_plus[r_i-2]+lambd_minus[r_i-2]) * coef - coef * lambd_minus[r_i-2]* alph1[r_i-2])
    g2_t_h[j+1][r_i-1] = (g_t_h[j][r_i-1] + 0.5*coef * lambd_minus[r_i-2] * beta1[r_i-2]) / (1 + 0.5*(lambd_plus[r_i-2]+lambd_minus[r_i-2]) * coef - 0.5*coef * lambd_minus[r_i-2]* alph1[r_i-2])

    for m in range(r_i - 2, -1, -1):#Ð²Ñ‹Ñ‡Ð¸ÑÐ»ÑÐµÐ¼ Ð¿Ñ€Ð¾Ð³Ð¾Ð½ÐºÐ¾Ð¹ g1_t_h Ð´Ð»Ñ ÑˆÐ°Ð³Ð° Ð¿Ð¾ Ð²Ñ€ÐµÐ¼ÐµÐ½Ð¸
        g1_t_h[j+1][m] = g1_t_h[j+1][m + 1] * alph1[m] + beta1[m]
        g2_t_h[j+1][m] = g2_t_h[j+1][m + 1] * alph2[m] + beta2[m]
    beta3[0] = g2_t_h[j+1][0]
    for i in range(1, r_i-1): #Ð²Ñ‹Ñ‡Ð¸ÑÐ»ÑÐµÐ¼ ÐºÐ¾ÑÑ„Ñ„Ð¸Ñ†Ð¸ÐµÐ½Ñ‚Ñ‹ Ð°Ð»ÑŒÑ„Ð° Ð¸ Ð±ÐµÑ‚Ð° Ð´Ð»Ñ g1
        alph3[i] = 0.5*C_p / (-0.5*C_p* alph3[i - 1] + 1 + C_p)
        beta3[i] = (g2_t_h[j+1][i] + 0.5*C_p * beta3[i-1]) / (-0.5*C_p * alph3[i-1] + 1 + C_p)
    g3_t_h[j+1][r_i-1] = (g2_t_h[j][r_i-1] + 0.5*coef * lambd_minus[r_i-2] * beta1[r_i-2]) / (1 + 0.5*(lambd_plus[r_i-2]+lambd_minus[r_i-2]) * coef - 0.5*coef * lambd_minus[r_i-2]* alph1[r_i-2])
    for m in range(r_i - 2, -1, -1):#Ð²Ñ‹Ñ‡Ð¸ÑÐ»ÑÐµÐ¼ Ð¿Ñ€Ð¾Ð³Ð¾Ð½ÐºÐ¾Ð¹ g1_t_h Ð´Ð»Ñ ÑˆÐ°Ð³Ð° Ð¿Ð¾ Ð²Ñ€ÐµÐ¼ÐµÐ½Ð¸
        g3_t_h[j+1][m] = g3_t_h[j+1][m + 1] * alph3[m] + beta3[m]
        g_t_h[j+1][m] = 2 * g3_t_h[j+1][m] - g1_t_h[j+1][m]
fig = plt.figure()
ax = plt.axes(xlim=(0, 1), ylim=(0, 2))
line = []
N = 2
for j in range(N):
    temp, = plt.plot([], [])
    line.append(temp)
line = tuple(line)
def init():
    for j in range(N):
        line[j].set_data([], [])
    return line,
def animate(i):
    #x_i = np.arange(0, L, h)
    #y = g_t_h[i]
    x_i = np.arange(0, L+h, h)
    y1 = g_t_h[i]
    ksi = x_i / (2 * np.sqrt(lambd * i * t))
    y2 = Temp * special.erfc(ksi)
    line[0].set_data(x_i, y1)
    line[1].set_data(x_i, y2)
    return line,
 
anim = FuncAnimation(fig, animate, init_func=init,
                               frames=20000, interval=10, blit=True, repeat = False)    
```
