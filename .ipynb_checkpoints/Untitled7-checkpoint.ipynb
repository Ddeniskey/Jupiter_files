```python
import numpy as np
import scipy
from scipy import special
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
plt.style.use('seaborn-pastel')
```

```python
#Ð¡Ñ…ÐµÐ¼Ð° 0 Ð´Ð»Ñ Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾Ð³Ð¾ ÑƒÑ€Ð°Ð²Ð½ÐµÐ½Ð¸Ñ Ð¿Ð»ÑŽÑÐ¾Ð¼ Ñ‚Ð¾Ñ‡Ð½Ð¾Ðµ Ñ€ÐµÑˆÐµÐ½Ð¸Ðµ Ð² Ð²Ð¸Ð´Ðµ Ñ„ÑƒÐ½ÐºÑ†Ð¸Ð¸ Ð¾ÑˆÐ¸Ð±Ð¾Ðº
lambd = 0.1
Temp=300
L = 1
T = 5
C_p = 0.4
m = 100
h = L / m
t = C_p * h ** 2 / lambd
n = int(T // t)
x_i = np.arange(0, L+h, h) # Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ Ð² ÑƒÐ·Ð»Ð°Ñ… Ð¿Ð¾ Ñ…
t_j = np.linspace(0, T, n) # Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ð² ÑƒÐ·Ð»Ð°Ñ… Ð¿Ð¾ t
r_j = len(t_j) # ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ ÑƒÐ·Ð»Ð¾Ð² Ð¿Ð¾ t
r_i = len(x_i) # ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ ÑƒÐ·Ð»Ð¾Ð² Ð¿Ð¾ x
g_t_h = np.zeros([r_j, r_i]) # Ð¸Ñ‚Ð¾Ð³Ð¾Ð²Ð°Ñ ÑÐµÑ‚ÐºÐ° Ñ€Ð°Ð·Ð¼ÐµÑ€Ð¾Ð¼ t_j*x_i
#Ð·Ð°Ð´Ð°Ð´Ð¸Ð¼ Ð³Ñ€Ð°Ð½Ð¸Ñ‡Ð½Ñ‹Ðµ ÑƒÑÐ»Ð¾Ð²Ð¸Ñ
def fi(x, Temp):
    ans = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] == 0:
            ans[i] = Temp
        else: 
            ans[i] = 0
    return(ans)
T_t_0 = fi(x_i, Temp) # Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ð¿Ñ€Ð¸ t = 0
T_x_0 = Temp # Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ð¿Ñ€Ð¸ x = 0
T_x_L = 0
g_t_h[0] = T_t_0 # Ð¸Ð½Ð¸Ñ†Ð¸Ð°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ ÑÐµÑ‚ÐºÐ¸ t = 0
for i in range(n):
    g_t_h[i][0] = T_x_0 # Ð³Ñ€Ð°Ð½Ð¸Ñ‡Ð½Ð¾Ðµ ÑƒÑÐ»Ð¾Ð²Ð¸Ðµ Ð¿Ñ€Ð¸ Ñ… =0
for j in range(r_j - 1):#Ð²Ñ€ÐµÐ¼Ñ
    for i in range(1, r_i - 1): #Ð¿Ñ€Ð¾ÑÑ‚Ñ€Ð°Ð½ÑÑ‚Ð²Ð¾
        g_t_h[j + 1][i] = C_p * g_t_h[j][i + 1] + (1 - 2 * C_p)*g_t_h[j][i] + C_p * g_t_h[j][i - 1]

fig = plt.figure()
ax = plt.axes(xlim=(0, 1), ylim=(0, 400))
line = []
N = 2
for j in range(N):
    temp, = plt.plot([], [])
    line.append(temp)
line = tuple(line)
def init():
    for j in range(N):
        line[j].set_data([], [])
    return line,
def animate(i):
    #x_i = np.arange(0, L, h)
    #y = g_t_h[i]
    x_i = np.arange(0, L+h, h)
    y1 = g_t_h[i]
    ksi = x_i / (2 * np.sqrt(lambd * i * t))
    y2 = Temp * special.erfc(ksi)
    line[0].set_data(x_i, y1)
    line[1].set_data(x_i, y2)
    return line,
 
anim = FuncAnimation(fig, animate, init_func=init,
                               frames=25000, interval=2, blit=True, repeat = False)
print(x_i)        
#Ð·ÐµÐ»ÐµÐ½Ð¾Ðµ-Ñ‚ÐµÐ¾Ñ€ÐµÑ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ðµ, ÑÐ¸Ð½ÐµÐµ - Ñ‡Ð¸ÑÐ»ÐµÐ½Ð½Ð¾Ðµ        
```
