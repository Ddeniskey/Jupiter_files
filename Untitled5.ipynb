```python
import pandas as pd
data = pd.read_csv('https://video.ittensive.com/python-advanced/data-5283-2019-10-04.utf.csv', sep=';')
unemp = pd.read_csv('https://video.ittensive.com/python-advanced/data-9753-2019-07-25.utf.csv', sep=';')
unemp = unemp.rename(columns={'global_id':'global_id', 'ID':'ID', 'Year':'Year', 'UnemployedMen':'UnemployedMen', 'UnemployedWomen':
                      'UnemployedWomen',
       'UnemployedYoung':'UnemployedYoung', 'UnemployedDisabled':'UnemployedDisabled', 'UnemployedTotal':'UnemployedTotal', 'Period':'Month',
       'Unnamed: 9':'Unnamed: 9'})
unemp.query('Year == 2015')
```

```python
data.query("AdmArea == 'Центральный административный округ'")
```

```python
data_all = pd.merge(data.query("AdmArea == 'Центральный административный округ'")
                    , unemp,on =['Year', 'Month'])
data_all.sort_values(by = 'Calls')
data_all.drop(labels=['global_id_x', 'ID_y'],axis=1)
```

```python
unemp1=unemp.groupby('Year').agg({'UnemployedMen':'mean',
                           'UnemployedWomen':'mean',
                           'UnemployedYoung':'mean',
                           'UnemployedDisabled':'mean',
                           'UnemployedTotal':'mean'})
unemp1['Disabled_percent']=unemp['UnemployedDisabled'] / unemp['UnemployedTotal'] * 100
unemp[unemp.Disabled_percent < 2.0]
```

```python
#Процент безработных в Москве
data_all1=data_all.groupby('Year', as_index=False).agg({'UnemployedMen':'mean',
                           'UnemployedWomen':'mean',
                           'UnemployedYoung':'mean',
                           'UnemployedDisabled':'mean',
                           'UnemployedTotal':'mean'})
data_all1['Disabled_percent']=data_all1['UnemployedDisabled'] / data_all1['UnemployedTotal'] * 100
data_all1
```

```python
import numpy as np
from  sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
unemp = unemp[unemp.Year >=2015]
x = np.array(unemp.Year).reshape(len(unemp.Year), 1)
y = np.array(unemp.Disabled_percent).reshape(len(unemp.Year), 1)
model = LinearRegression()
model.fit(x, y)
plt.scatter(x, y, color='orange')
x = np.append(x, [2020]).reshape(len(unemp.Year)+1, 1)
plt.plot(x, model.predict(x))
print(model.predict(np.array(2020).reshape(1,1)))
```
